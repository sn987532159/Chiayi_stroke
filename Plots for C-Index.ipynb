```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_df = pd.read_excel("model results.xlsx", index_col=0)
```

```python
plt.rcdefaults() 
```

```python
#plt.style.use('classic')
plt.figure(figsize=(6, 3), dpi=300)
plt.plot(results_df["Lasso-cvd"], label = "Penalized Cox model", linestyle = ":", marker = "o", alpha = 0.7, c='r')
plt.plot(results_df["DeepSurv-cvd"], label = "DeepSurv", linestyle = ":", marker = "p", alpha = 0.7, c='g')
plt.plot(results_df["RSF-cvd"], label = "RSF", linestyle = ":", marker = "*", alpha = 0.7, c='b')
plt.scatter(25, results_df["cox-cvd"][25], label = "Multivariate Cox model", marker = "x", color="black", alpha = 1, s=50)
plt.legend(loc = 4)
plt.xlabel("Number of variables", labelpad= 5, fontweight='bold', size = 12)
plt.ylabel("C-index", rotation = 90, loc = "center", labelpad= 5, fontweight='bold', size = 12)
plt.title("Cardiovascular death")
plt.xticks(results_df.index, size = 8)
plt.yticks(size = 8)
plt.xlim(0.5, 25.5)
plt.grid(alpha = 0.3, linestyle = '-.')
plt.legend(loc=4, prop={'size': 6})
#plt.savefig("All-cause_cvd.png", dpi = 300, transparent = False, bbox_inches='tight')
plt.show()
```
